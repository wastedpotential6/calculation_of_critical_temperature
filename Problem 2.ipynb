```python
import pandas as pd
import sklearn
import numpy as np
```

```python
df_train = pd.read_csv('train.csv')
df_train
```

```python
formula_train = pd.read_csv('formula_train.csv')
formula_train
```

```python
df_test = pd.read_csv('test.csv')
df_test
```

```python
formula_test = pd.read_csv('formula_test.csv')
formula_test
```

```python
X_drop = formula_train.drop(['material','critical_temp','He','Ne','Ar','Kr','Xe'],axis = 1) #17000x81
temp1 = df_train.drop(['critical_temperature'],axis = 1) #17000x81
X_prod_train = np.multiply(X_drop,temp1)
X_prod_train
```

```python
X = X_prod_train.astype('int')
X_list_train = list(X.itertuples(index = False))
X_list_train
```

```python
y = formula_train['critical_temp']
y = y.astype('int')
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 2)

#reg = linear_model.LinearRegression()
#reg.fit(X_list_train,y)
```

```python
X_poly = poly.fit_transform(X_list_train)
X_poly.shape
```

```python
poly.fit(X_poly,y)
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_poly,y)
```

```python
linreg.score(X_poly,y)
```

```python
linreg.coef_.shape
```

```python
X_drop_test = formula_test.drop(['material','He','Ne','Ar','Kr','Xe'],axis = 1) #17000x81
temp2 = df_test #17000x81
X_prod_test = np.multiply(X_drop_test,temp2)
X_prod_test.shape
```

```python
X_prod_test = X_prod_test.astype('int')
X_list_test = list(X_prod_test.itertuples(index = False))
X_list_test
```

```python
X_poly_answer = poly.fit_transform(X_list_test)
X_poly_answer.shape
```

```python
answer = linreg.predict(X_poly_answer)
answer
```
